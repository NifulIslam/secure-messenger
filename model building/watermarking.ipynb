```python
import numpy as np 
import pandas as pd
import cv2
from matplotlib import pyplot as plt
```

```python
img= cv2.imread("../input/flickr-image-dataset/flickr30k_images/flickr30k_images/1002674143.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img=cv2.resize(img, (500,400))
plt.imshow(img)
```

```python
img.shape
```

```python
# watermark
watermark= cv2.imread("../input/uiu-logo/UIU-Logo-250.png")
watermark = cv2.cvtColor(watermark, cv2.COLOR_BGR2RGB)
watermark=cv2.resize(watermark, (500,400))
plt.imshow(watermark)
plt.title("watermark")
t=5
```

```python
# image with watermark will have 6 bits for image and 2 bits for watermark
img_free_2=(img>>2)<<2
watermark_free_2=(watermark>>6)
img_with_watermark=cv2.add(img_free_2,watermark_free_2)
plt.imshow(img_with_watermark)
plt.title("image with watermark")
t=5
```

# reconstructed watermark

```python
regenerated_watermark=(img_with_watermark<<6)
plt.imshow(regenerated_watermark)
```
